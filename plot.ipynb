```python
import numpy as np
import matplotlib.pyplot as plt

# load npy data and plot
data1 = np.load('metrics/res_50/mean_IoU_list.npy')
data2 = np.load('metrics/res_50_pretrain/mean_IoU_list.npy')
data3 = np.load('metrics/res_101/mean_IoU_list.npy')
data4 = np.load('metrics/res_101_pretrain/mean_IoU_list.npy')

# plt.plot(data)
# plot 4 lines in a figure and legend
plt.plot(data1, label='ResNet50')
plt.plot(data2, label='ResNet50_pretrain')
plt.plot(data3, label='ResNet101')
plt.plot(data4, label='ResNet101_pretrain')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('mean IoU')
plt.title('Mean IoU')
plt.show()
```
